```python
custAge=data["custAge"]
q25, q75= np.percentile(custAge,[25,75])
IQR=q75-q25
q0=q25-(IQR*1.5)
q100=q75+(IQR*1.5)
print("q0",q0)
print("q100",q100)
filter=custAge [ (custAge<=q100) & (custAge>=q0) ]
plt.boxplot(filter)
plt.grid()
plt.show()
```

```python
for i in numname:
    col=data[i]
    q25, q75= np.percentile(col,[25,75])
    IQR=q75-q25
    q0=q25-(IQR*1.5)
    q100=q75+(IQR*1.5)
    print(i)
    print("q0: ",q0)
    print("q25: ",q25)
    print("q75: ",q75)  
    print("q100: ",q100)
    filter=col [ (col>=q0) & (col<=q100) ]
    plt.boxplot(filter) 
    plt.grid()
    plt.show()
```

```python
for i in numname:
    col = data[i]
    q25, q75 = np.percentile(col, [25, 75])
    IQR = q75 - q25
    q0 = q25 - (IQR * 1.5)
    q100 = q75 + (IQR * 1.5)
    print(i)
    print("q0: ", q0)
    print("q25: ", q25)
    print("q75: ", q75)  
    print("q100: ", q100)
    col[(col <= q0) & (col >= q100)] = np.NaN
    filter=col
    plt.boxplot(filter)  
    plt.grid()
    plt.show()
```
